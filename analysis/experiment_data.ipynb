# Experiment Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/experiment_data/plot.csv")
df["Beta_Acc"] *= 10
df["Beta"] *= 100
df["Spine"] *= 5
print(df.head())

df.plot()                
plt.xlabel("Index")    
plt.ylabel("Value")
plt.title("Sensor Data")
plt.tight_layout()
plt.show()
```
